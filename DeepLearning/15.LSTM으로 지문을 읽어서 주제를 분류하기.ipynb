```python
import pandas as pd
df = pd.read_csv('data/lstm.csv')
df.head()
```

```python
df.category.unique()
```

```python
# 데이터에 사용된 중복 없는 전체 단어 갯수 파악
results = set()
df['paragraph'].str.lower().str.split().apply(results.update)
vocab_size = len(results)
print(vocab_size)
```

```python
# 단어를 숫자로 인코딩
paragraphs = df['paragraph'].to_list()
paragraphs[0:2]
```

```python
from tensorflow import keras
keras.utils.set_random_seed(1) # 여기서도 set seed 할 수 있음
```

```python
# one hot encoding
encoded_paragraphs = [keras.preprocessing.text.one_hot(p, vocab_size) for p in paragraphs]
encoded_paragraphs
```

```python
# 데이터에서 가장 긴 문장의 단어 갯수 파악
max_len = max([len(p) for p in encoded_paragraphs])
print(max_len)
```

```python
model = keras.Sequential()

# 문맥 생성 단계
model.add(keras.layers.Embedding(vocab_size, 5, input_length=max_len))
# Embedding: 원핫인코딩 레이어
# 5: 
  
```
