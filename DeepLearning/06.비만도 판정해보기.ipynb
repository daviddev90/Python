### 정규화

```python
# weight, height 정규화
csv['weight'] = csv['weight'] / csv['weight'].max()
csv['height'] = csv['height'] / csv['height'].max()
```

```python
csv.head()
```

```python
# label
bclass = {'thin': [1, 0, 0], 'normal': [0, 1, 0], 'fat': [0, 0, 1]}
y = []
for i, v in enumerate(csv['label']):
    y.append(bclass[v])
y
```

```python
# Training
X_train = csv[['weight', 'height']][0:15000]
y_train = y[0:15000]
```

```python
# Test
X_test = csv[['weight', 'height']][15001:20001]
y_test = y[15001:20001]
```

---
# Deep Learning 만들기

```python
from tensorflow import keras
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(units=512, input_shape=(2,), activation='relu'))
model.add(keras.layers.Dense(units=3, activation='softmax'))
```

```python
from pickletools import optimize
from sklearn import metrics

model.compile(
    optimizer='rmsprop', # sgd
    metrics='accuracy',
    loss='categorical_crossentropy' # 정수가 아니고 데이터가 표준화되어서
)
```

```python
# 데이터 훈련하기
import numpy as np

model.fit(
  np.array(X_train),
  np.array(y_train),
  epochs=50,
)
```

```python
# 테스트 데이터로 평가하기
score = model.evaluate(X_test,  np.array(y_test))
print('loss', score[0])
print('accuracy', score[1])
```
