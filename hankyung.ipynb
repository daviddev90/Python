```python
# scrape from xml
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time

# import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
```

```python
url = 'https://hankyung.com/newssitemap-4.xml'
# get xml
res = requests.get(url)
soup = BeautifulSoup(res.text, 'lxml')
# get all url
urls = [url.text for url in soup.find_all('loc')]
# get all publication_date
publication_date = [date.text for date in soup.find_all('news:publication_date')]

# to dataframe
df = pd.DataFrame({'url': urls, 'publication_date': publication_date})

df
```

```python
import urllib.request as req

innerurl = df.iloc[0, 0]
print(innerurl)

# # get innerurl with selenium
# chrome_options = webdriver.ChromeOptions()
# driver = webdriver.Chrome(ChromeDriverManager().install(), options=chrome_options)
# driver.get(innerurl)
# time.sleep(1)
# #parse innerurl
# innerurl_soup = BeautifulSoup(driver.page_source, 'lxml')
# # print text
# innerurl_soup.text

# request with headers(useragent) to get html
res = req.Request(innerurl, headers={'User-Agent': 'Mozilla/5.0'})
innerurl_html = req.urlopen(res).read()
innerurl_soup = BeautifulSoup(innerurl_html, 'html')
innerurl_soup
```
