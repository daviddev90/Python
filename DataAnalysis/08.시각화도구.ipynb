```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("data/시도별 전출입 인구수.xlsx")
df.head()
```

```python
# NaN을 전 (Prev) 데이터로 채우기
df = df.fillna(method= 'ffill')
df
```

```python
df['전출지별']
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
# 전출지별이 서울특별시 이고 전입지별이 서울특별시가 아닌 곳
df_seoul = df.loc[( df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시') ]
df_seoul.head()
```

```python
# 전출지별 컬럼 삭제
df_seoul =  df_seoul.drop('전출지별' , axis=1)
df_seoul
```

```python
# 전입지별 -> 전입지 로 컬럼이름 변경
df_seoul.rename(columns={'전입지별':'전입지'} ,inplace=True)
df_seoul
```

```python
# df_seoul의 index를 전입지로 변경 
df_seoul.set_index('전입지' , inplace=True)
df_seoul
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택
sr_one = df_seoul.loc['경기도']
sr_one.head()
```

```python
plt.plot( sr_one.index, sr_one.values)
plt.show()
```

```python
plt.plot(sr_one)
plt.show()
```

> 한글폰트

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
plt.plot(sr_one)

# 차트 제목 추가
plt.title('서울 -> 경기 인구 이동')

# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동 인구수')
  
plt.show()
```

```python
# 그래프 꾸미기
# 그림 사이즈 지정하기(가로 14인치 , 세로 5인치)
plt.figure(figsize =(14,5))

# x축 눈금 라벨 회전하기
plt.xticks(rotation = 45) # rotation = 90 or 'vertical' / 원하는 각도 수치
# y축 눈금 라벨 회전하기
plt.yticks(rotation = 60) 

plt.title('서울->경기 인구 이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.plot(sr_one)

# 범례 표시 
plt.legend(labels = ['서울->경기'] , loc = 'best') # loc(location) best 최적의 자리에 맞춰줌 

plt.show()
```

> 부산 -> 서울 인구이동 그래프 그리기

```python
# 부산에서 전출 , 서울로 전입 
df_busan = df.loc[( df['전출지별'] == '부산광역시') & (df['전입지별'] == '서울특별시') ]
df_busan.drop('전출지별' , axis=1 ,  inplace=True)
df_busan.set_index('전입지별' , inplace=True)
df_busan = df_busan.T
```

```python
# 그래프 꾸미기
# 그림 사이즈 지정하기(가로 14인치 , 세로 5인치)
plt.figure(figsize =(14,5))

# x축 눈금 라벨 회전하기
plt.xticks(rotation = 45) # rotation = 90 or 'vertical' / 원하는 각도 수치
# y축 눈금 라벨 회전하기
plt.yticks(rotation = 60) 

plt.title('부산->서울 인구 이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.plot(df_busan)

# 범례 표시 
plt.legend(labels = ['부산 -> 서울'] , loc = 'best') # loc(location) best 최적의 자리에 맞춰줌 

plt.show()
```

### 그래프에 Style 적용하기

```python
plt.figure(figsize =(14,5))
plt.xticks(rotation = 45) 
plt.yticks(rotation = 60) 
plt.title('서울->경기 인구 이동' , size = 15)
plt.xlabel('기간', size = 15)
plt.ylabel('이동 인구수', size = 15)


# Style적용
plt.style.use('ggplot')

# Marker표시
plt.plot(sr_one , marker = "o" , markersize = 10)

# y축 범위 지정
plt.ylim(50000, 800000)

# 화살표 표시
plt.annotate('꼬부기!!!!!', xy=(20, 620000), xytext=(2, 290000), arrowprops=dict(
    arrowstyle='->', color='skyblue', lw=5), xycoords='data', textcoords='data')

plt.annotate('인구이동 증가(1970 ~ 1995)', # 텍스트
            xy= (11, 400000), # 텍스트 위치 기준점
            rotation=26, # 텍스트 회전각도
            size=15, # 텍스트 크기
            va='baseline', # 텍스트 위치
            ha='center') # 텍스트 정렬


# 화살표 표시
plt.annotate('', xy=(46, 450000), xytext=(28, 600000), arrowprops=dict(
    arrowstyle='->', color='darkgreen', lw=5), xycoords='data', textcoords='data')

plt.annotate('인구이동 감소(1995 ~ 2007)', # 텍스트
            xy= (37, 520000), # 텍스트 위치 기준점
            rotation=350, # 텍스트 회전각도
            size=15, # 텍스트 크기
            va='baseline', # 텍스트 위치
            ha='center') # 텍스트 정렬

# 범례 폰트사이즈 
plt.legend(labels = ['서울->경기'] , loc = 'best' , fontsize = 20 ) 

# 이미지로 저장하기
plt.savefig("data/seoul.png")

plt.show()
```

# 화면을 분할하여 그래프 여러개 그리기

```python
# Warning 무시하기
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.style.use('ggplot')

# 그래프 객체 생성(figure에 2개의 서브플롯을 생성)
fig = plt.figure(figsize =(10,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(sr_one , "o" , markersize = 10)
ax2.plot(sr_one , marker = "o" , markerfacecolor='green', markersize = 10, color='olive', linewidth=2, label='서울->경기')
ax2.legend(loc='best')

ax1.set_ylim(50000, 800000)
ax2.set_ylim(50000, 800000)

ax1.set_xticklabels(sr_one.index, rotation = 75)
ax2.set_xticklabels(sr_one.index, rotation = 75)

plt.show()
```

```python
# 서울에서 '충청남도', '경상북도', '강원도'로 이동한 인구 이동 그래프
data = df_seoul.loc[['충청남도', '경상북도', '강원도']]
data.head()
plt.style.use('ggplot')

fig = plt.figure(figsize =(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(data.columns, data.loc['충청남도'], marker = "o" , markersize = 10, color='red', linewidth=2, label='서울 > 충청남도')
ax.plot(data.columns, data.loc['경상북도'], marker = "o" , markersize = 10, color='blue', linewidth=2, label='서울 > 경상북도')
ax.plot(data.columns, data.loc['강원도'], marker = "o" , markersize = 10, color='green', linewidth=2, label='서울 > 강원도')

ax.legend(loc='best')
ax.set_title('서울에서 충남, 경북, 강원도로 인구 이동', size = 17)

ax.set_xlabel('기간', size = 16)
ax.set_ylabel('이동 인구수', size = 16)

ax.set_xticklabels(data.columns, rotation = 75, size = 12)


plt.show()
```

```python
# 2 X 2  그래프 그리기
fig = plt.figure(figsize =(20,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

data4 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도']]

ax1.plot(data4.columns, data4.loc['충청남도'], marker = "o" , markersize = 10, color='green', linewidth=2, label='서울 > 충청남도')
ax2.plot(data4.columns, data4.loc['경상북도'], marker = "o" , markersize = 10, color='blue', linewidth=2, label='서울 > 경상북도')
ax3.plot(data4.columns, data4.loc['강원도'], marker = "o" , markersize = 10, color='red', linewidth=2, label='서울 > 강원도')
ax4.plot(data4.columns, data4.loc['전라남도'], marker = "o" , markersize = 10, color='yellow', linewidth=2, label='서울 > 전라남도')

ax1.set_title('서울 > 충청남도 인구 이동', size = 15)
ax2.set_title('서울 > 경상북도 인구 이동', size = 15)
ax3.set_title('서울 > 강원도 인구 이동', size = 15)
ax4.set_title('서울 > 전라남도 인구 이동', size = 15)

ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best')

ax1.set_xticklabels(data.columns, rotation=75, size=9)
ax2.set_xticklabels(data.columns, rotation=75, size=9)
ax3.set_xticklabels(data.columns, rotation=75, size=9)
ax4.set_xticklabels(data.columns, rotation=75, size=9)


ax1.set_ylim(10000, 70000)
ax2.set_ylim(10000, 70000)
ax3.set_ylim(10000, 70000)
ax4.set_ylim(10000, 70000)

plt.show()
```

```python
df_seoul_to_busan = df.loc[(df['전출지별'] == '서울특별시') & (df['전입지별'] == '부산광역시')]
df_busan_to_seoul = df.loc[(df['전출지별'] == '부산광역시') & (df['전입지별'] == '서울특별시')]

# drop first two columns
df_seoul_to_busan.drop(df_seoul_to_busan.columns[0:2], axis=1, inplace=True)
df_busan_to_seoul.drop(df_busan_to_seoul.columns[0:2], axis=1, inplace=True)

plt.style.use('ggplot')

fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(df_seoul_to_busan.T, marker="o",
        markersize=10, color='red', linewidth=2, label='서울 > 충청남도')
ax.plot(df_busan_to_seoul.T, marker="o",
        markersize=10, color='blue', linewidth=2, label='부산 > 충청남도')

# x축 라벨
ax.set_xlabel('기간', size = 16, color='blue')
# y축 라벨
ax.set_ylabel('이동 인구수', size = 16, color='blue')

# 제목
ax.set_title('서울에서 부산으로 인구 이동', size=20, color='purple')

# 범례
ax.legend(loc='best')

# x축 눈금 간격
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

### 2010 ~ 2017년의 기간동안 서울에서 각 시도로 이동한 인구의 합계를 구하여 그래프로 비교

```python
col_years = list(map(str, range(2010, 2018)))
col_years

df_4 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_years]
df_4.columns = col_years

# 합계 column 추가
df_4['합계'] = df_4.sum(axis=1)

# 정렬하기
df_4_total = df_4[['합계']].sort_values(by='합계', ascending=True)

# 합계 column 제거
df_4.drop('합계', axis=1, inplace=True)

plt.style.use('ggplot')

df_4_total.plot(kind='barh', color='cornflowerblue', figsize=(10, 5), title='서울에서 타 시도로 인구 이동', grid=False, rot=0, width=0.8)
```
