```python
테마주_0705 = 테마주
# 테마주_0705에 '어제종가1','그제종가1', '어제거래량1','그제거래량1', '어제종가2', '그제종가2', '어제거래량2','그제거래량2' column 추가
테마주_0705['어제종가1'] = 0
테마주_0705['그제종가1'] = 0
테마주_0705['어제종가2'] = 0
테마주_0705['그제종가2'] = 0
# first row, '종목코드1' col
for i in range(len(테마주_0705)):
  종목코드1 = str(테마주_0705.iloc[i, 2])
  종목코드2 = str(테마주_0705.iloc[i, 3])
  # get 종목코드1 row in stock0705
  어제종가1 = stock0705.loc[종목코드1, '종가']
  그제종가1 = stock0704.loc[종목코드1, '종가']
  어제종가2 = stock0705.loc[종목코드2, '종가']
  그제종가2 = stock0704.loc[종목코드2, '종가']
  테마주_0705.iloc[i, 4] = 어제종가1
  테마주_0705.iloc[i, 5] = 그제종가1
  테마주_0705.iloc[i, 6] = 어제종가2
  테마주_0705.iloc[i, 7] = 그제종가2

# save to csv
테마주_0705.to_csv('테마주_0705.csv')
테마주_0705
```

```python
테마주_0705['등락율1'] = (테마주_0705['어제종가1'] / 테마주_0705['그제종가1'] - 1) * 100
테마주_0705['등락율2'] = (테마주_0705['어제종가2'] / 테마주_0705['그제종가2'] - 1) * 100
테마주_0705['평균등락율'] = (테마주_0705['등락율1'] + 테마주_0705['등락율2']) / 2

테마주_0705.head()
```

```python
# 평균등락율 순 정렬
테마주_0705 = 테마주_0705.sort_values(by='평균등락율', ascending=False)
테마주_0705.head(15)
```
