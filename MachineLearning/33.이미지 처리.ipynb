# Numpy를 Image로 변환하기

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
fruits = np.load('data/fruits.npy')
print(fruits.shape)
```

```python
# Numpy를 image로 저장하기
from PIL import Image

# 300개 이미지 loop
for i in range(300):
  image = fruits[i]
  image = Image.fromarray(image).convert('L')
  # 03d format jpeg로 저장
  image.save('data/fruits/{:03d}.jpg'.format(i), 'JPEG')
```

# image를 불러와 numpy 배열로 변환하기

```python
import glob
```

```python
numberOfData = len(glob.glob('data/fruits/*.jpg'))
imgSize = 100
```

```python
# 배열 초기화
fruitsData = np.zeros(numberOfData*imgSize*imgSize, dtype=np.int32)
# 사진갯수 X 100 X 100 으로 reshape
fruitsData = fruitsData.reshape(numberOfData, imgSize, imgSize)
```

```python
# 사진 읽어오기
i = 0
for file in sorted(glob.glob('data/fruits/*.jpg')):
  image = Image.open(file)
  image = np.array(image, dtype=np.int32)
  # 이미지 데이터 저장
  fruitsData[i] = image
  i += 1
```

```python
# Numpy Image Data를 File로 저장하기
np.save('data/fruitsData.npy', fruitsData) #.npy는 빼도 됨
```
