# 머신러닝이란?
: 수많은 데이터를 학습시켜 패턴(y = ax + b)을 찾아 이를 기반으로 데이터를 분류하거나 예측
- 일반 소프트웨어(결정론적 소프트웨어)는 언제나 똑같은 과정을 거쳐 똑같은 결과를 내보냄
- 머신러닝 기반 소프트웨어(비 결정론적 소프트웨어)는 데이터와 알고리즘(Ctree 등)에 따라 때로 다른 결과를 추론해 냄
- 추론: 이미 알고 있는 것으로부터 논리적 결론을 도출하는 것

### 결정론적 소프트웨어와 비결정론적 소프트웨어의 차이점
: 예를 들어 지문인식기 소프트웨어를 결정론적 소프트웨어로 처리하면? - 지문 입력기에 지문이 조금만 틀어져도 인식 불가(입력한 지문과 완전히 같아야 함)   
이에 반해 비결정론적 소프트웨어는 지문이 조금 틀어져도 추론을 통해 지문의 일치여부를 결정

### 머신러닝 개요
: 머신러닝이란 인공지능 연구 과제 중의 하나로, 인간의 뇌가 자연스럽게 수행하는 '학습'이라는 능력을 컴퓨터로 구현하는 방법
- 어느 정도 규모가 있는 샘플 데이터를 입력으로 넣어 분석
- 입력한 데이터에서 일정한 규칙을 찾아냄(찾아내는 알고리즘이 많음)
- 찾아낸 규칙을 기반으로 다른 데이터를 분류(classification)하거나 예측(regression)

## 머신러닝의 종류
##### 1)지도학습(Supervised Learning)
- 학습 시 데이터와 답을 같이 학습한 후 다른 데이터의 값을 예측
##### 2)비지도학습(Unsupervised Learning)
- 학습 시 데이터만 학습한 후 다른 데이터의 규칙성을 찾음. 본질적인 구조 등을 확인할 때 사용
##### 3)강화학습(Reinforcement Learning)
- 부분적으로 답을 입력한 후 데이터를 기반으로 최적의 답을 찾아 냄
##### 4)딥러닝(Deep Learning)
- 비지도학습과 비지도학습의 관련이 있으나, 특이성은 학습자료에 대한 인간의 개입을 받지 않는다.

### 데이터 분석가 입장에서
- 딥러닝이 제일 쉬움(분석가가 할 게 없음)
- 강화학습이 그 다음(분석가가 할 게 적은 편)
- 지도학습이 제일 어려움(다 셋팅해줘야 함)

# 머신러닝의 흐름
(1) 데이터 수집   
(2) 데이터 가공   
(3) 데이터 학습   
&nbsp;&nbsp;&nbsp;(3-1) 학습 방법 선택,   
&nbsp;&nbsp;&nbsp;(3-2) 매개 변수 조정   
&nbsp;&nbsp;&nbsp;(3-3) 모델 학습    
(4) 모델 평가   
(5) 예측력이 좋은 경우 다음 단계 진행하고 아니면 (3)번 부터 다시   
(6) 성공


# 머신러닝 응용분야
### 클래스 분류(Classification)
- 특정 데이터에 레이블을 붙여 분류
- 스팸 메일 분류, 필기인식, 증권 사기 방지 등

### 클러스터링(Clustering): 그룹 나누기
- 값의 유사성을 기반으로 데이터를 여러 그룹으로 나눔
- 사용자의 취향을 그룹으로 묶어 취향에 맞는 광고 제공

### 추천(Recommendation)
- 특정 데이터를 기반으로 다른 데이터 추천
- 인터넷 서점에서 구매한 책들을 기반으로 다른 책을 추천
### 회귀

### 차원 축소(Dimensionality Reduction)
- 데이터의 특성을 유지하면서 데이터의 양을 줄이는 것
- 데이터를 시각화하거나 구조를 추출해서 용량을 줄여 계산을 빠르게 하거나 메모리를 절약할 때 사용

# 프로젝트 과정
: 문제정의 => 데이터 확윽 => 모델 구현 => 검증 => 테스트
### (1)문제정의
- 프로젝트의 목표를 정의하고 그에 따른 계획을 세우는 단계
- 대략적으로 어떤 데이터를 수집해서 어떤 머신러닝 알고리즘으로 문제을 해결할 지 결정하는 단계
### (2) 데이터 획득
- 모델 학습에 필요한 데이터 수집
- 성공적인 머신러닝 프로젝트는 머신러닝 실력보다도 얼마나 충분히 데이터를 획득했는냐에 의해 결정
- 획득한 데이터는 학습데이터, 검증데이터, 테스트 데이터로 구분해 사용
